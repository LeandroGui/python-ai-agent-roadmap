En el nivel anterior, us√°bamos variables sueltas (`agent_name`, `is_active`) para definir a Cortex. Esto se vuelve inmanejable si queremos tener dos agentes o si el sistema crece.

## 1. La Clase (`class`): El Plano
Una clase es una plantilla o "plano de arquitectura". Define **qu√©** atributos y capacidades tendr√° nuestro agente, pero no es el agente en s√≠.

## 2. El Objeto (Instancia): La Construcci√≥n
Cuando usamos el plano para crear una entidad real en la memoria, eso es un Objeto.
`my_agent = Agent("Cortex")`

## 3. El Constructor (`__init__`) y `self`
* `__init__`: Es el primer m√©todo que se ejecuta al crear el objeto. Aqu√≠ configuramos el estado inicial.
* `self`: Es una referencia a "m√≠ mismo". Permite que el c√≥digo sepa que `self.name` se refiere al nombre de *este* agente espec√≠fico y no de otro.

```python
# --- Definici√≥n de la Clase (El Plano) ---
class Agent:
    """
    Clase base para un Agente de IA.
    Define el estado y comportamiento gen√©rico.
    """

    # El Constructor: Configura el estado inicial
    def __init__(self, name: str, version: float = 1.0):
        self.name = name          # Atributo: Identidad
        self.version = version    # Atributo: Configuraci√≥n
        self.energy = 100         # Atributo: Estado din√°mico
        self.memory = []          # Atributo: Almacenamiento

    # M√©todo: Comportamiento (Acci√≥n)
    def perceive(self, input_data: str):
        """Recibe datos del exterior y gasta energ√≠a."""
        print(f"[{self.name}] Percibiendo: '{input_data}'")
        self.energy -= 5
        self.memory.append(input_data)

    def status_report(self):
        """Devuelve el estado actual."""
        return f"{self.name} v{self.version} | Energ√≠a: {self.energy}% | Memorias: {len(self.memory)}"

# --- Instanciaci√≥n (Creando Objetos Reales) ---

# Creamos al agente "Cortex" usando el plano
cortex = Agent("Cortex", 2.0)

# Creamos otro agente diferente usando el mismo plano
legion = Agent("Legion", 0.5)

# Interactuamos con los objetos
cortex.perceive("Hola mundo")
legion.perceive("Iniciando protocolo")

print("--- REPORTE DE ESTADO ---")
print(cortex.status_report())
print(legion.status_report())
```
