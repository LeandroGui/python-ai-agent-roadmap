## 1. Variables y Tipos de Datos: El Estado del Agente
En Python, una **variable** es un espacio en la memoria de la computadora donde guardamos informaci칩n. Para un Agente, las variables definen su **Estado Interno** (qui칠n es, qu칠 versi칩n tiene, si est치 encendido o apagado).

A diferencia de otros lenguajes, Python tiene **tipado din치mico** (no necesitas declarar el tipo expl칤citamente), pero en Ingenier칤a de IA usamos **Type Hints** (pistas de tipo) para que nuestro c칩digo sea robusto y entendible por otros sistemas.

* `str` (String): Texto. Usado para el nombre del agente o comandos.
* `float`: Decimales. Usado para m칠tricas de precisi칩n o confianza (0.0 a 1.0).
* `bool`: Verdadero/Falso. Usado para estados binarios (Activo/Inactivo).

```python
# --- CELDA 1: Variables con Sem치ntica de Agente ---
# En Python normal: x = 10
# En Cortex: Usamos nombres descriptivos y Type Hints (preparaci칩n para el futuro)

agent_name: str = "Cortex_Genesis"
version: float = 0.1
is_active: bool = True
confidence_level: float = 0.95

print(f"Sistema: {agent_name} v{version}")
print(f"Estado Activo: {is_active}")
```

## 2. Condicionales: La Toma de Decisiones
Un agente no es 칰til si siempre hace lo mismo. Necesita reaccionar a est칤mulos externos (Inputs).

Las estructuras de control de flujo (`if`, `elif`, `else`) son la forma m치s primitiva de "inteligencia". Permiten que el c칩digo se bifurque:
1.  **Percibir:** El agente lee el `user_input`.
2.  **Evaluar:** Compara la entrada con reglas predefinidas.
3.  **Actuar:** Ejecuta el bloque de c칩digo correspondiente.

> **Nota de Dise침o:** Los agentes modernos usan Modelos de Lenguaje (LLMs) para decidir, pero incluso ellos necesitan reglas l칩gicas "duras" (Hardcoded constraints) para seguridad (ej: "si el usuario pide algo peligroso, detenerse").

```python
# --- CELDA 2: L칩gica de Decisi칩n (El "if" es el razonamiento b치sico) ---
user_input = "analizar datos"

# Un agente primitivo usa reglas fijas (Hardcoded logic)
if user_input == "apagar":
    is_active = False
    print("Apagando sistemas...")
elif "analizar" in user_input:
    # Simulaci칩n de pensamiento
    if confidence_level > 0.8:
        print(f"Entendido. Iniciando an치lisis con confianza del {confidence_level * 100}%")
    else:
        print("Confianza insuficiente para ejecutar esta tarea.")
else:
    print("Comando no reconocido.")
```
