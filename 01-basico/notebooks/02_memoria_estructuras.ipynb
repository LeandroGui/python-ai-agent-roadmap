## 1. Listas: Memoria Secuencial (Historial)
Una **Lista** (`list`) en Python es una colecci贸n ordenada y mutable de elementos.
En el contexto de un Agente, las listas son perfectas para:
* **Logs de auditor铆a:** Registrar paso a paso qu茅 hizo el sistema.
* **Historial de chat:** Guardar la conversaci贸n en orden cronol贸gico (`[mensaje1, mensaje2, ...]`).

Usamos bucles (`for`) para recorrer estas listas, lo que equivale a que el agente "revise su memoria" para buscar patrones o resumir eventos pasados.

```python
# --- CELDA 1: Listas (Memoria Secuencial / Historial) ---
# Un agente necesita recordar qu茅 pas贸 antes.
command_history = []

# Simulamos flujo de entrada
command_history.append("usuario: hola")
command_history.append("agente: hola, 驴en qu茅 puedo ayudar?")
command_history.append("usuario: resume este texto")

print(f"ltimo comando: {command_history[-1]}")
print(f"Total de interacciones: {len(command_history)}")

# Iteraci贸n (Procesamiento en lote)
print("--- Log de Conversaci贸n ---")
for event in command_history:
    print(f">> {event}")
```
